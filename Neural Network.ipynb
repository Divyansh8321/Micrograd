```python
class Module: 
    
    def zero_grad(self):  # Used for changing the gradients back to zero after every backprop iteration
        for p in self.parameters():
            p.grad = 0 
    
    def parameters(self): # Just providing common interface
        return []
    
    def step(self, optimizer):
        for p in self.parameters():
            optimizer.step(p)
```

```python
class Neuron(Module):
    def __init__(self, nin, normalization=None):
        self.w = [Value(np.random.uniform(-1, 1)) for _ in range(nin)]
        self.b = Value(np.random.uniform(-1, 1))
        self.normalization = normalization

    def __call__(self, x):
        weighted_inputs = [wi * xi for wi, xi in zip(self.w, x)]
        out = sum(weighted_inputs, self.b)

        if self.normalization:
            out = self.normalization.forward(out)

        return out

    def parameters(self):
        return self.w + [self.b]

    def __repr__(self):
        return f'dense neuron with ({len(self.w) + 1} parameters)'
    
```

```python
class Layer(Module):
    def __init__(self, nin, nout, normalization=None):
        self.neurons = [Neuron(nin, normalization=normalization) for _ in range(nout)]

    def __call__(self, x):
        out = [neuron(x) for neuron in self.neurons]
        return out[0] if len(out) == 1 else out  # Just to get the single output from the list

    def parameters(self):
        return [p for neuron in self.neurons for p in neuron.parameters()]

    def __repr__(self):
        return f'layer of {len(self.neurons)}'
```

```python
class MLP(Module):
    def __init__(self, nin, nouts, normalization=None):
        size = [nin] + nouts
        self.layers = [Layer(size[i], size[i + 1], normalization=normalization) for i in range(len(nouts))]

    def __call__(self, x):
        for layer in self.layers:
            x = layer(x)
        return x

    def parameters(self):
        return [p for layer in self.layers for p in layer.parameters()]

    def __repr__(self):
        return f'layer of {len(self.layers)}'
```

```python
# Adding Optimizers for more applications

class SGD:
    def __init__(self, learning_rate=0.001):
        self.learning_rate = learning_rate

    def step(self, p):  # In gradient descent, gradient is a vector pointed in the direction of increased los
        p.data -= self.learning_rate * p.grad
        
        
class Adam:
    def __init__(self, learning_rate=0.001, beta1=0.9, beta2=0.999, epsilon=1e-8):
        self.learning_rate = learning_rate
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.t = 0
        self.m = 0
        self.v = 0

    def step(self, p):
        self.t += 1
        self.m = self.beta1 * self.m + (1 - self.beta1) * p.grad
        self.v = self.beta2 * self.v + (1 - self.beta2) * p.grad**2
        m_hat = self.m / (1 - self.beta1**self.t)
        v_hat = self.v / (1 - self.beta2**self.t)
        p.data -= self.learning_rate * m_hat / (v_hat**0.5 + self.epsilon)

class RMSprop:
    def __init__(self, learning_rate=0.001, decay_rate=0.9, epsilon=1e-8):
        self.learning_rate = learning_rate
        self.decay_rate = decay_rate
        self.epsilon = epsilon
        self.accumulated_square = 0

    def step(self, p):
        self.accumulated_square = self.decay_rate * self.accumulated_square + (1 - self.decay_rate) * p.grad**2
        p.data -= self.learning_rate * p.grad / (self.accumulated_square + self.epsilon)**0.5
```

```python
# Adding Normalisation Techniques

class BatchNormalization:
    def __init__(self, epsilon=1e-5, momentum=0.9):
        self.epsilon = epsilon
        self.momentum = momentum
        self.running_mean = 0
        self.running_var = 1
        self.gamma = Value(1.0)
        self.beta = Value(0.0)

    def forward(self, x):
        if not hasattr(self, 'train_mode') or self.train_mode:
            mean = np.mean(x, axis=0)
            var = np.var(x, axis=0)
            self.running_mean = self.momentum * self.running_mean + (1 - self.momentum) * mean
            self.running_var = self.momentum * self.running_var + (1 - self.momentum) * var
            x_hat = (x - mean) / (var + self.epsilon)**0.5
        else:
            x_hat = (x - self.running_mean) / (self.running_var + self.epsilon)**0.5

        return self.gamma * x_hat + self.beta

    def parameters(self):
        return [self.gamma, self.beta]
    
    
class LayerNormalization:
    def __init__(self, epsilon=1e-5):
        self.epsilon = epsilon
        self.gamma = Value(1.0)
        self.beta = Value(0.0)

    def forward(self, x):
        mean = np.mean(x, axis=1, keepdims=True)
        var = np.var(x, axis=1, keepdims=True)
        x_hat = (x - mean) / (var + self.epsilon)**0.5
        return self.gamma * x_hat + self.beta

    def parameters(self):
        return [self.gamma, self.beta]
```
